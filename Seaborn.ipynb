##1. DISTPLOT (Distribution Plot) sebaran nilainya
ðŸ§  Intinya:

Distplot dipakai untuk melihat sebaran satu variabel numerik secara visual â€” mirip histogram yang ditambah densitas kurva.
---



```python
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# contoh pakai dataset "tips"
tips = sns.load_dataset("tips")

tips.head()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

##ðŸ’¡ Penjelasan logika:

total_bill = variabel angka (misal jumlah tagihan)

histogram menunjukkan berapa sering nilai muncul

kde=True = menambahkan kurva smooth untuk lihat bentuk distribusi

Tujuannya: kita bisa lihat data:
âœ” nilai tengahnya di mana
âœ” miring ke kanan/kiri atau normal
âœ” ada puncak banyak datanya

##ðŸ“ˆ 2. JOINTPLOT hubungan antara nilai total_bill dan tip
ðŸ§  Intinya:

Jointplot dipakai untuk dua variabel â€” kita lihat hubungan keduanya plus sebaran masing-masing.

```python
sns.jointplot(data=tips, x="total_bill", y="tip")
plt.show()
```

total_bill = sumbu X

tip = sumbu Y

Plot utama = titik-titik yang menunjukkan hubungan dua variabel

Di atas & di kanan ada histogram untuk masing-masing variabel

Berguna untuk:
âœ” melihat apakah dua variabel itu berkorelasi
âœ” melihat sebaran sendiri-sendiri

##ðŸ”— 3. PAIRPLOT hubungan semua sekaligus
ðŸ§  Intinya:

Pairplot pakai ketika ada banyak variabel numerik dan kita mau lihat hubungan antar semua kombinasi.

```python
sns.pairplot(data=tips)
plt.show()
```

##ðŸ’¡ Penjelasan logika:

Kalau dataset punya kolom A, B, C, D â†’ pairplot bikin grid:
âœ” A vs B
âœ” A vs C
âœ” B vs C
dan seterusnya

Di diagonal biasanya ada histogram untuk masing-masing kolom

Ini sangat berguna di awal analisis untuk tahu mana variabel yang berkaitan satu sama lain

## 4. HEATMAP

Dipakai buat:

ðŸ‘‰ melihat KORELASI
ðŸ‘‰ lihat aja garis yg ada di kanan, kalau makin ke bawah makin lemah

```python
# hitung hubungan semua kolom
heat = tips.corr(numeric_only=True)


```

```python
# tampilkan hubungan itu dalam bentuk warna
sns.heatmap(heat.corr());
```

## LOGIKA

Heatmap dipakai untuk:

ðŸ‘‰ melihat hubungan antar angka secara cepat
ðŸ‘‰ mana yang saling berpengaruh kuat, mana yang tidak

Secara teknis dia nampilin correlation (korelasi).

## 5. Countplot

Dipakai untuk menghitung berapa banyak tiap kategori muncul di dataset.

```python
sns.countplot(data=tips, x="sex")
plt.show()
```

##Logika:

x="sex" â†’ hitung berapa baris untuk masing-masing nilai sex

Tampilnya bar chart otomatis
â†’ berguna kalau kita mau lihat distribusi kategori.

## 6.Barplot

Mirip countplot, tapi bisa menunjukkan ***rata-rata atau statistik lain*** dari sebuah nilai berdasarkan kategori.

```python
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
```

##Logika:

x="day" = kategori hari

y="total_bill" = nilai yang mau diringkas

Barplot otomatis hitung mean (rata-rata) dari total_bill tiap day
â†’ berguna untuk lihat â€œberapa rata-rata tagihan per hari?â€.

## 7. Boxplot

Ini visual statistik yang memperlihatkan ***quartile dan outlier*** dari distribusi nilai.

```python
sns.boxplot(data=tips, x="day", y="tip")
plt.show()
```

##Logika:

Kotak menunjukkan 25%â€“75% data (IQR)

Garis tengah = median

Garis luar (whiskers) = rentang normal data

Titik jauh = outlier
â†’ berguna untuk bandingkan penyebaran tip per hari.

## 8. Violinplot

Mirip boxplot + distribusi dalam satu grafik.

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.show()
```

##Logika:

Bentuk lebar di tengah = banyak data

Mirip KDE di kiri & kanan
â†’ lebih detail daripada boxplot biasa karena menunjukkan bentuk distribusi.

```python
sns.barplot(data=tips, x="day", y="tip")
plt.title("Rata-rata Tip per Hari")
plt.show()
```
